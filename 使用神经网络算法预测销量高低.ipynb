# 使用神经网络算法预测销量高低

```python
import pandas as pd
inputfile ='sales_data.xls'
data =pd.read_excel(inputfile ,index_col ='序号')
```

```python
data
```

```python
# 编码
data[data=='好']=1
data[data=='是']=1
data[data=='高']=1
data[data !=1]=0
```

```python
# 构造训练集
x=data.iloc[:,:3]#.as_matrix()#.astype(int)
y=data.iloc[:,3].as_matrix().astype(int)
```

```python
from  keras.models import Sequential
from keras.layers.core import Dense ,Activation,Dropout
```

```python
model =Sequential()#建立模型
model.add(Dense(units=10, activation='relu', input_dim=3))
model.add(Dropout(0.5))
model.add(Dense(units=1, activation='sigmoid'))
model.compile(loss= 'binary_crossentropy',optimizer ='adam')#,class_models ='binary')
#添加Dropout使得部分网络节点休眠
```

```python
model.fit(x,y,nb_epoch =2000,batch_size =10)
yp =model.predict_classes(x).reshape(len(y))
```

```python
yp
```

```python
from sklearn import metrics
metrics.confusion_matrix(y,yp)
```
